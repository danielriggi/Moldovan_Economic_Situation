```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
df_mda = pd.read_csv('../data/all_indicators_mda.csv', index_col=0)
```

```python
df_mda.head()
```

```python
df_mda = df_mda.iloc[1:,:]
```

```python
df_mda.head()
```

```python
df_mda['Value'] = pd.to_numeric(df_mda['Value'], errors = 'coerce')
```

```python
df_mda.dropna(subset=['Value'], inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
table = df_mda[df_mda['Indicator Name'] == 'GDP per capita, PPP (current international $)']
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlabel('Year', fontsize=20)
plt.ylabel('GDP Per Capita', fontsize=20)
plt.title('Moldova GDP Per Capita', fontsize=25)

x = table['Year'].sort_values()
y = table['Value'].sort_values()
ax.xaxis.set_tick_params(labelsize=20, labelrotation=90.0)
ax.yaxis.set_tick_params(labelsize=20, )
ax.bar(x,y, color='Blue')
```
